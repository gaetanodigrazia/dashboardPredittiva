### Valori mancanti in percentuale per ogni colonna nel Dataframe

```python
def msv_1(data, thresh = 20, color = 'black', edgecolor = 'black', height = 3, width = 15):
    
    plt.figure(figsize = (width, height))
    percentage = (data.isnull().mean()) * 100
    percentage.sort_values(ascending = False).plot.bar(color = color, edgecolor = edgecolor)
    plt.axhline(y = thresh, color = 'r', linestyle = '-')
    
    plt.title('Missing values percentage per column', fontsize=20, weight='bold' )
    
    plt.text(len(data.isnull().sum()/len(data))/1.7, thresh+12.5, f'Columns with more than {thresh}% missing values', fontsize=12, color='crimson',
         ha='left' ,va='top')
    plt.text(len(data.isnull().sum()/len(data))/1.7, thresh - 5, f'Columns with less than {thresh}% missing values', fontsize=12, color='green',
         ha='left' ,va='top')
    plt.xlabel('Columns', size=15, weight='bold')
    plt.ylabel('Missing values percentage')
    plt.yticks(weight ='bold')
    
    return plt.show()
msv_1(train_df, 20, color=sns.color_palette('Reds',15))
```

### Right-Side Skewed

Il termine "Right-Side Skewed" si riferisce a una distribuzione di dati in cui la maggior parte dei valori si trovano sulla parte sinistra della distribuzione e pochi valori si trovano sulla parte destra della distribuzione. In altre parole, la coda della distribuzione si estende verso destra.

Questo tipo di distribuzione può anche essere chiamato "skewed to the right" o "positively skewed". Ad esempio, se si considera una distribuzione dei redditi delle famiglie, ci si aspetterebbe di vedere molti valori bassi (vicino alla mediana) e pochi valori molto elevati (vicino alla coda destra della distribuzione).

Cosa comporta la right-side skewed?
La presenza di una distribuzione "Right-Side Skewed" può comportare alcune implicazioni nell'analisi dei dati, tra cui:

- La media può non essere una misura rappresentativa della tendenza centrale dei dati. A causa della presenza di valori estremi nella coda destra della distribuzione, la media può essere influenzata in modo significativo da questi valori. In questo caso, la mediana potrebbe essere una scelta migliore per rappresentare la tendenza centrale dei dati.
- La presenza di una coda lunga sulla destra della distribuzione può portare a una maggior incidenza di valori anomali rispetto a una distribuzione simmetrica. Ciò potrebbe richiedere un'analisi ulteriore per determinare se questi valori anomali sono effettivamente errori nei dati o rappresentano un fenomeno interessante da esplorare ulteriormente.
- La distribuzione "Right-Side Skewed" può influenzare le analisi statistiche che si basano sull'assunzione di normalità dei dati, come la regressione lineare. In questo caso, potrebbe essere necessario considerare l'utilizzo di metodi statistici non parametrici che non richiedono l'assunzione di normalità dei dati.

```python
indexers = [StringIndexer(inputCol=column, outputCol=column+'_index',
                          handleInvalid='keep').fit(train_df) for column in train_string_columns]

pipeline = Pipeline(stages=indexers)
```

#### In questo passaggio vediamo la funzione pipeline in azione. Prima eseguiamo una stima (fit) e poi una trasformazione.

```python
train_indexed = pipeline.fit(train_df).transform(train_df)
```

#### Per mostrare il risultato della nostra pipeline, eseguiamo la print del numero di colonne.

```python
print(len(train_indexed.columns))
```

#### Ripetiamo le operazioni per i dati di test.

```python
test_string_columns = []

for col, dtype in test_df.dtypes:
    if dtype == 'string':
        test_string_columns.append(col)

indexers2 = [StringIndexer(inputCol=column, outputCol=column+'_index',
                           handleInvalid='keep').fit(test_df) for column in test_string_columns]

pipeline2 = Pipeline(stages=indexers2)
test_indexed = pipeline2.fit(test_df).transform(test_df)

print(len(test_indexed.columns))
```

#### Questa funzione mi permette di costruire una colonna vettoriale che non tiene conto del SalePrice.
#### Ha come output delle fetures e setta gli "invalid data = keep" per mantenerli.

```python
vectorAssembler = VectorAssembler(inputCols=train_indexed.drop("SalePrice").columns, outputCol='features').setHandleInvalid("keep")    

train_vector = vectorAssembler.transform(train_indexed)
```

#### Ripetiamo l'operazione per i dati di test.

```python
vectorAssembler2 = VectorAssembler(inputCols=test_indexed.columns, outputCol='features').setHandleInvalid("keep")

test_vector = vectorAssembler2.transform(test_indexed)
```

#### Impostiamo nei dati di test il saleprice a 0 perchè dobbiamo predirlo e quindi non deve essere utilizzato.

```python
test_vector = test_vector.withColumn("SalePrice", lit(0))
```

#### Train-test split.

```python
splits = train_vector.randomSplit([0.8, 0.2])
train = splits[0]
val = splits[1]
```

#### Creazione del modello.

```python
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol='features', labelCol='SalePrice', maxIter=10, regParam=0.8, elasticNetParam=0.1)
lr_model = lr.fit(train)

trainingSummary = lr_model.summary


lr_predictions = lr_model.transform(val)
```

#### Mostriamo le prime 5 prediction.

```python
lr_predictions.select("prediction", "SalePrice", "features").show(5)
```

#### Prediction.

```python
lr_evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="SalePrice", metricName="r2")
r2_percentage = "{:.0%}".format(trainingSummary.r2)
print("R2: " + r2_percentage)
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
```

```python
rf = RandomForestRegressor(featuresCol='features', labelCol='SalePrice',
                           maxDepth=20,
                           minInstancesPerNode=2,
                           bootstrap=True
                           )
rf_model = rf.fit(train)

rf_predictions = rf_model.transform(val)
rf_predictions.select("prediction", "SalePrice", "features").show(5)

r2_evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="SalePrice", metricName="r2")
r2_percentage = "{:.0%}".format(r2_evaluator.evaluate(rf_predictions))
rmse_evaluator = RegressionEvaluator(predictionCol="prediction", labelCol="SalePrice", metricName="rmse")
print("R2: " + r2_percentage)
print("RMSE: %f" % rmse_evaluator.evaluate(rf_predictions))
```

#### Creazione di un modello RandomForest preliminare.

```python
rf = RandomForestRegressor(labelCol="SalePrice", featuresCol="features")
```

#### Valutazione del modello.

```python
rfevaluator = RegressionEvaluator(predictionCol="prediction", labelCol="SalePrice", metricName="rmse")
```

#### Creazione del ParamGrid per la Cross Validation.

```python
rfparamGrid = (ParamGridBuilder()
             .addGrid(rf.maxDepth, [2, 5, 10, 20, 30])
             .addGrid(rf.maxBins, [10, 20, 40])
             .addGrid(rf.numTrees, [5, 20, 50])  
             .build())
```

```python
cost = np.zeros(20)
for k in range(2,20):
    kmeans = KMeans()\
            .setK(k)\
            .setSeed(1) \
            .setFeaturesCol("features")\
            .setPredictionCol("prediction")

    model = kmeans.fit(train)
    cost[k] = model.summary.trainingCost # requires Spark 2.0 or later
```

```python
fig, ax = plt.subplots(1,1, figsize =(8,6))
ax.plot(range(2,20),cost[2:20])
ax.set_xlabel('k')
ax.set_ylabel('cost')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

#### Creazione di CrossValidator a 6-fold.

```python
rfcv = CrossValidator(estimator = rf,
                      estimatorParamMaps = rfparamGrid,
                      evaluator = rfevaluator,
                      numFolds = 6)
```

#### Esecuzione della Cross Validation.

```python
rfcvModel = rfcv.fit(train)
print(rfcvModel)
```

#### Utilizziamo il test set per misurare l'accuratezza del modello su nuovi dati.

```python
rfpredictions = rfcvModel.transform(val)
```
